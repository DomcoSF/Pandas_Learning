```python
import pandas as pd
```

```python
df_item_pedido = pd.read_csv('/content/item_pedido.csv')
df_produto = pd.read_csv('/content/produto.csv')
df_pedido = pd.read_csv('/content/pedido.csv')
```

```python
df_item_pedido.merge(df_produto, how ='left', on=['descItem'])
```

```python
df_join = df_item_pedido.merge(right=df_produto
                               , how='left'
                               )

df_join.merge(df_pedido, how='left')
```

```python
df_join = (df_item_pedido.merge(right=df_produto, how='left')
              .merge(df_pedido, how='left'))

df_join
```
