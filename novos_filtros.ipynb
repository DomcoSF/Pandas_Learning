```python
import pandas as pd
```

```python
df_produto = pd.read_csv('/content/produto.csv')
df_produto
```

```python
df_produto['primeiroNome'] = df_produto['descItem'].apply(lambda x: x.split(" ")[0])
df_produto
```

```python
df_produto = df_produto.sort_values(
    by=['vlPreco','descItem'],
    ascending=[False,True],
)

df_produto.drop_duplicates(subset=['primeiroNome','vlPreco'],keep='first')
df_produto
```
